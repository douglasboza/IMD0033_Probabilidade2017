```python
## Douglas Pinheiro Boza

import math
import matplotlib.pyplot as plt
import seaborn as sns


outcome_counts = [x * 0.1 for x in range(0, 16)]

def find_tmr(t, l):
    return ( 3*(2.718281828)**(-2*l*t) - 2*(2.718281828)**(-3*l*t))

def find_stmr(t, l):
    return ( (3/2)*(2.718281828**(-l*t)) ) - ((1/2)*(2.718281828**(-3*l*t)))

def find_sr(t, l):
    return 2.718281828**(-l*t)

plt.figure(figsize=(10, 6))

outcome_probs = [find_tmr(t, 5) for t in outcome_counts]
plt.plot(outcome_counts, outcome_probs, label="TMR")

outcome_probs = [find_sr(5, t) for t in outcome_counts]
plt.plot(outcome_counts, outcome_probs, label='Sem redundância')

outcome_probs = [find_stmr(5, t) for t in outcome_counts]
plt.plot(outcome_counts, outcome_probs, label="TMR/Simplex")

plt.ylabel("Confiabilidade")
plt.xlabel("Tempo")
plt.title("ƛ = 5")
plt.legend()
plt.show()

#===========

outcome_probs = [find_tmr(t, 5) for t in outcome_counts]
plt.plot(outcome_counts, outcome_probs, label="TMR")

outcome_probs = [find_sr(5, t) for t in outcome_counts]
plt.plot(outcome_counts, outcome_probs, label='Sem redundância')

outcome_probs = [find_stmr(5, t) for t in outcome_counts]
plt.plot(outcome_counts, outcome_probs, label="TMR/Simplex")

plt.ylabel("Confiabilidade")
plt.xlabel("Tempo")
plt.title("ƛ = 5")
plt.legend()
plt.show()

#===========

outcome_probs = [find_tmr(t, 5) for t in outcome_counts]
plt.plot(outcome_counts, outcome_probs, label="TMR")

outcome_probs = [find_sr(5, t) for t in outcome_counts]
plt.plot(outcome_counts, outcome_probs, label='Sem redundância')

outcome_probs = [find_stmr(5, t) for t in outcome_counts]
plt.plot(outcome_counts, outcome_probs, label="TMR/Simplex")

plt.ylabel("Confiabilidade")
plt.xlabel("Tempo")
plt.title("ƛ = 5")
plt.legend()
plt.show()
```

```python

outcome_counts = list(range(10))

outcome_counts = list(range(10))

def find_cdf(k, l):
    sum = 0
    for a in range(0, k):
        sum = sum + ((l**a)/math.factorial(a)) * 2.71828**(-l)
    return sum

outcome_probs = [find_cdf(i, 1) for i in outcome_counts]
sns.set()
plt.title('lambda = 1')
plt.bar(outcome_counts, outcome_probs)
plt.show()

outcome_probs = [find_cdf(i, 4) for i in outcome_counts]
sns.set()
plt.title('lambda = 4')
plt.bar(outcome_counts, outcome_probs)
plt.show()

outcome_probs = [find_cdf(i, 10) for i in outcome_counts]
sns.set()
plt.title('lambda = 10')
plt.bar(outcome_counts, outcome_probs)
plt.show()
```

```python

outcome_counts = list(range(20))

def find_hype(k, n, d, m):
    com1 = math.factorial(d)/(math.factorial(k)*math.factorial(math.fabs(d-k)))
    com2 = math.factorial(n-d)/(math.factorial(math.fabs(m-k))*math.factorial((n-d)-(m-k)))
    com3 = math.factorial(n)/(math.factorial(m)*math.factorial(n-m))
    
    return com1*com2/com3

outcome_probs = [find_hype(k,100,5,4) for k in outcome_counts]
Y = outcome_probs
X = outcome_counts
plt.scatter(X,Y, alpha=1, s=70, color="blue")
plt.title('Distribuição Hipergeométrica - PDF')

plt.show()

plt.show()
```

```python

outcome_counts = list(range(20))

def find_hypecdf(k, n, d, m):
    com4 = 0
    for a in range(0, k):
        com1 = math.factorial(d)/(math.factorial(a)*math.factorial(math.fabs(d-a)))
        com2 = math.factorial(n-d)/(math.factorial(math.fabs(m-a))*math.factorial((n-d)-(m-a)))
        com3 = math.factorial(n)/(math.factorial(m)*math.factorial(n-m))
        com4 = com4 + (com1*com2/com3)
    return com4

outcome_probs = [find_hypecdf(k,100,5,4) for k in outcome_counts]
Y = outcome_probs
X = outcome_counts
plt.scatter(X,Y, alpha=1, s=70, color="blue")
plt.title('Distribuição Hipergeométrica - CDF')
plt.show()

plt.show()
```
