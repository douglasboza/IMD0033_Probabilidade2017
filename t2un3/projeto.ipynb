```python
## Douglas Pinheiro Boza

import math
import matplotlib.pyplot as plt
import seaborn as sns


outcome_counts = [x * 0.1 for x in range(0, 16)]

def find_tmr(t, l):
    return ( 3*(2.718281828)**(-2*l*t) - 2*(2.718281828)**(-3*l*t))

def find_tmr2(t, l):
    return ( 2*((2.718281828)**(-l*t)) - (2.718281828)**(-2*l*t))

#def find_tmr4(t, l):
 #   return ( 5*((2.718281828)**(-4*l*t)) - 4*(2.718281828)**(-5*l*t))


def find_mttf(outcome_counts, outcome_probs):
    mttf=0;
    for x in range(0, 16):
        mttf = mttf+outcome_probs[x]*0.1
    return mttf
    
def find_stmr(t, l):
    return ( (3/2)*(2.718281828**(-l*t)) ) - ((1/2)*(2.718281828**(-3*l*t)))

def find_sr(t, l):
    return 2.718281828**(-l*t)

plt.figure(figsize=(15, 6))

outcome_probs = [find_tmr(t, 5) for t in outcome_counts]
print(find_mttf(outcome_counts, outcome_probs))
plt.plot(outcome_counts, outcome_probs, label="TMR", color='red')
plt.axvline(x=find_mttf(outcome_counts, outcome_probs), color='red')

outcome_probs = [find_sr(t, 5) for t in outcome_counts]
#print(find_mttf(outcome_counts, outcome_probs))
plt.plot(outcome_counts, outcome_probs, label='Sem redundância', color='blue')
plt.axvline(x=find_mttf(outcome_counts, outcome_probs) , color='blue')

outcome_probs = [find_stmr(t, 5) for t in outcome_counts]
#print(find_mttf(outcome_counts, outcome_probs))
plt.plot(outcome_counts, outcome_probs, label="TMR/Simplex", color='#cecece')
plt.axvline(x=find_mttf(outcome_counts, outcome_probs), color='#cecece')


outcome_probs = [find_tmr2(t, 5) for t in outcome_counts]
#print(find_mttf(outcome_counts, outcome_probs))
plt.plot(outcome_counts, outcome_probs, label="TMR/2", color='#FFE15F')
plt.axvline(x=find_mttf(outcome_counts, outcome_probs), color='#FFE15F')


#outcome_probs = [find_tmr4(t, 5) for t in outcome_counts]
#print(find_mttf(outcome_counts, outcome_probs))
#plt.plot(outcome_counts, outcome_probs, label="TMR/4", color='#222222')
#plt.axvline(x=find_mttf(outcome_counts, outcome_probs), color='#222222')


plt.ylabel("Confiabilidade")
plt.xlabel("Tempo")
plt.title("ƛ = 5")
plt.legend()
plt.show()

#===========
plt.figure(figsize=(15, 6))

outcome_probs = [find_tmr(t, 2) for t in outcome_counts]
#print(find_mttf(outcome_counts, outcome_probs))
plt.plot(outcome_counts, outcome_probs, label="TMR", color='red')
plt.axvline(x=find_mttf(outcome_counts, outcome_probs), color='red')

outcome_probs = [find_sr(t, 2) for t in outcome_counts]
#print(find_mttf(outcome_counts, outcome_probs))
plt.plot(outcome_counts, outcome_probs, label='Sem redundância', color='blue')
plt.axvline(x=find_mttf(outcome_counts, outcome_probs) , color='blue')

outcome_probs = [find_stmr(t, 2) for t in outcome_counts]
#print(find_mttf(outcome_counts, outcome_probs))
plt.plot(outcome_counts, outcome_probs, label="TMR/Simplex", color='#cecece')
plt.axvline(x=find_mttf(outcome_counts, outcome_probs), color='#cecece')


outcome_probs = [find_tmr2(t, 2) for t in outcome_counts]
#print(find_mttf(outcome_counts, outcome_probs))
plt.plot(outcome_counts, outcome_probs, label="TMR/2", color='#FFE15F')
plt.axvline(x=find_mttf(outcome_counts, outcome_probs), color='#FFE15F')


#outcome_probs = [find_tmr4(t, 2) for t in outcome_counts]
#print(find_mttf(outcome_counts, outcome_probs))
#plt.plot(outcome_counts, outcome_probs, label="TMR/4", color='#222222')
#plt.axvline(x=find_mttf(outcome_counts, outcome_probs), color='#222222')


plt.ylabel("Confiabilidade")
plt.xlabel("Tempo")
plt.title("ƛ = 2")
plt.legend()
plt.show()


#===========
plt.figure(figsize=(15, 6))

outcome_probs = [find_tmr(t, 15) for t in outcome_counts]
#print(find_mttf(outcome_counts, outcome_probs))
plt.plot(outcome_counts, outcome_probs, label="TMR", color='red')
plt.axvline(x=find_mttf(outcome_counts, outcome_probs), color='red')

outcome_probs = [find_sr(t, 15) for t in outcome_counts]
#print(find_mttf(outcome_counts, outcome_probs))
plt.plot(outcome_counts, outcome_probs, label='Sem redundância', color='blue')
plt.axvline(x=find_mttf(outcome_counts, outcome_probs) , color='blue')

outcome_probs = [find_stmr(t, 15) for t in outcome_counts]
#print(find_mttf(outcome_counts, outcome_probs))
plt.plot(outcome_counts, outcome_probs, label="TMR/Simplex", color='#cecece')
plt.axvline(x=find_mttf(outcome_counts, outcome_probs), color='#cecece')


outcome_probs = [find_tmr2(t, 15) for t in outcome_counts]
#print(find_mttf(outcome_counts, outcome_probs))
plt.plot(outcome_counts, outcome_probs, label="TMR/2", color='#FFE15F')
plt.axvline(x=find_mttf(outcome_counts, outcome_probs), color='#FFE15F')


#outcome_probs = [find_tmr4(t, 15) for t in outcome_counts]
#print(find_mttf(outcome_counts, outcome_probs))
#plt.plot(outcome_counts, outcome_probs, label="TMR/4", color='#222222')
#plt.axvline(x=find_mttf(outcome_counts, outcome_probs), color='#222222')


plt.ylabel("Confiabilidade")
plt.xlabel("Tempo")
plt.title("ƛ = 15")
plt.legend()
plt.show()
```
