````python
#1-

import pandas as pd
turma20171 = pd.read_csv('turmas-20171.csv', sep=";")

turma20171_imd1 = turma20171.drop_duplicates(['codigo'])
turma20171_imd1 = turma20171_imd1[turma20171_imd1['codigo'].str.contains("IMD")]
turma20171_imd1 = turma20171_imd1[turma20171_imd1['situacao_turma'].str.contains("CONSOLIDADA")]

print("Total = ", len(turma20171_imd1))
print( turma20171_imd1["nome_ascii"] )

````

````python
#2
# Realize um estudo envolvendo as propriedades de assimétria e curtose para as turmas 
# IMD0033 (probabilidade) consolidadas em 2017.1. Utilize gráficos e explique os resultados. 
# Reforce também os horários das disciplinas nos resultados. 

from scipy.stats import skew
from scipy.stats import kurtosis
import seaborn as sns
import matplotlib.pyplot as plt


matricula20171 = pd.read_csv('matricula-componente-20171.csv', sep=";")
matricula20171_uni = matricula20171.drop_duplicates(['id_turma', 'id_discente'])

id_probabilidade = turma20171[turma20171["codigo"].str.contains("IMD0033")]
id_probabilidade = id_probabilidade.drop_duplicates("id_turma")["id_turma"]

imd0033_uni = matricula20171_uni.loc[matricula20171_uni['id_turma'].isin(id_probabilidade)]
imd0033_uni_not_null = imd0033_uni.loc[imd0033_uni['media_final'] >=0]

turmas_probabilidade = imd0033_uni_not_null["id_turma"].unique()


fig2, axes2 = plt.subplots(nrows=len(turmas_probabilidade), ncols=1, figsize=(10,20))
sns.set()

count=0

for i in turmas_probabilidade:
    turma_vez = imd0033_uni_not_null[imd0033_uni_not_null["id_turma"] == i]
    horario = turma20171[turma20171["id_turma"] == i]["descricao_horario"].values
    sns.distplot(turma_vez['media_final'], kde = True, ax=axes2[count])
    axes2[count].title.set_text(horario+ ' | skew = '+str(round(skew(turma_vez["media_final"]), 2))+ ' | kurtosis = '+str(round(kurtosis(turma_vez["media_final"]),2)) )    
    count+=1
plt.show()    
````

````python
#3
# Utilize a biblioteca seaborn para traçar gráficos que descrevam as distribuições estatísticas 
# (utilizar as funções facetgrid e map) de todas as turmas IMD0030 (Linguagem de Programação I) 
# consolidas em 2017.1. Na análise, assuma como referência as médias finais e a situação dos alunos. 


id_progr = turma20171[turma20171["codigo"].str.contains("IMD0030")]  
id_progr = id_progr.drop_duplicates("id_turma")["id_turma"]

mat20171_imd0030 = matricula20171_uni.loc[matricula20171_uni['id_turma'].isin(id_progr)]

mat20171_imd0030_notnull = mat20171_imd0030.loc[mat20171_imd0030['media_final'] >=0]
mat20171_imd0030_notnull = mat20171_imd0030_notnull.drop_duplicates("id_discente")

sns.set()
fig, (ax3, ax31) = plt.subplots(figsize=(25,8), ncols=1, nrows=2)

ax3.set_title("Média final IMD0030")
sns.distplot(mat20171_imd0030_notnull['media_final'], kde = False, ax=ax3, bins=30)

ax31.set_title("Situação IMD0030")
sns.countplot(mat20171_imd0030_notnull['situacao'], ax=ax31)

plt.subplots_adjust(top=2)

plt.show()
````

````python
#4
# Utilize a biblioteca seaborn para traçar gráficos de histograma referente as médias finais de todas 
# as turmas IMD0029 (Estrutura de dados I) consolidadas em 2017.1. Incremente nos gráficos linhas que 
# descrevam a média e mediana da distribuição. 

matricula20171 = pd.read_csv('matricula-componente-20171.csv', sep=";")
matricula20171_uni = matricula20171.drop_duplicates(['id_turma', 'id_discente'])

id_estrutura = turma20171[turma20171["codigo"].str.contains("IMD0029")]
id_estrutura = id_estrutura.drop_duplicates("id_turma")["id_turma"]

imd0029_uni = matricula20171_uni.loc[matricula20171_uni['id_turma'].isin(id_estrutura)]
imd0029_uni_not_null = imd0029_uni.loc[imd0029_uni['media_final'] >=0]

turmas_estrutura = imd0029_uni_not_null["id_turma"].unique()

fig2, axes2 = plt.subplots(nrows=len(turmas_estrutura), ncols=1, figsize=(10,20))

sns.set()

count=0


for i in turmas_estrutura:
    turma_vez = imd0029_uni_not_null[imd0029_uni_not_null["id_turma"] == i]
    horario = turma20171[turma20171["id_turma"] == i]["descricao_horario"].values
    sns.distplot(turma_vez['media_final'], kde = True, ax=axes2[count])
    axes2[count].title.set_text(horario+ ' | skew = '+str(round(skew(turma_vez["media_final"]), 2))+ ' | kurtosis = '+str(round(kurtosis(turma_vez["media_final"]),2)) )    
    axes2[count].axvline(turma_vez['media_final'].median(), color="green")
    axes2[count].axvline(turma_vez['media_final'].mean(), color="red")
    plt.suptitle("Mean (red) vs Median (green)")
    count+=1
plt.show()    
````
